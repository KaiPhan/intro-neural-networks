The loss function `nn.NLLLoss(reduction='sum')` utilizes the parameter `reduction='sum'` to compute the total loss over the entire batch instead of the mean (`'mean'`). The choice between `'sum'` and `'mean'` depends on specific needs:

- `'sum'`: When using `reduction='sum'`, the loss function sums the losses of each sample in the batch. This is useful when direct control over the optimization process is needed, as it is independent of the batch size. However, adjusting the learning rate is crucial, as the total loss can increase rapidly with a large batch size.

- `'mean'`: When using `reduction='mean'`, the loss function calculates the average loss per sample in the batch. This is useful when you want your model not to depend on the batch size. However, if the batch size varies significantly between epochs, using `'mean'` can lead to a reduction in the effective learning rate for the model.

In this case, `reduction='sum'` is used because we want to calculate the total loss over the entire batch, and then divide by the number of samples in the dataset to compute the average loss (see the lines `train_loss = tot_train_loss / len(trainloader.dataset)` and `test_loss = tot_test_loss / len(testloader.dataset)`). This ensures that the average loss is independent of the batch size. If we were using `reduction='mean'` instead of `'sum'`, we would need to remove the division by the batch size when calculating the average loss.

### Why `reduction='mean'` is not influenced by the `batch_size`?  
In the context of optimization, using `reduction='mean'` does not directly depend on the batch size for computing the loss. However, the effective impact of the learning rate is influenced by the batch size. If the batch size increases, the gradient values used for updating the weights during optimization are averaged over more samples, potentially leading to a more stable estimate of the gradient. Therefore, adjusting the learning rate is crucial when working with different batch sizes to ensure effective training.

## Note:  
The training loss is not a percentage, but a measure of how well the model fits the training data. A lower training loss means the model makes fewer errors on the training data, but it does not necessarily mean it performs better on new data.  
The test loss and accuracy are better indicators of how well the model generalizes to new data. The test accuracy is a percentage that shows how many of the test images the model predicted correctly.  
For example, a test accuracy of 0.837 means the model correctly classified 83.7% of the test images.


1. **Using `.sum().item()` for Accuracy:**
   ```python
   equals = top_class == labels.view(*top_class.shape)
   accuracy += equals.sum().item()
   ```
   In this approach, you calculate the number of correct predictions by summing the boolean values in the `equals` tensor (which contains `True` for correct predictions and `False` for incorrect predictions). The `.sum().item()` converts the sum to a Python scalar, which is then added to the `accuracy` variable.  
   
**For example:**
```python
equals = [False, True, True]
accuracy += equals.sum().item()
```
The `equals.sum()` operation sums boolean values (considering `True` as 1, `False` as 0). So, `equals.sum()` is 2, and `accuracy` is incremented by 2.

2. **Using `torch.mean(equals.type(torch.FloatTensor))` for Accuracy:**
   ```python
   equals = top_class == labels.view(*top_class.shape)
   accuracy += torch.mean(equals.type(torch.FloatTensor))
   ```
   In this approach, you calculate the accuracy by first converting the boolean tensor `equals` to a float tensor using `equals.type(torch.FloatTensor)`. This tensor has 1.0 for correct predictions and 0.0 for incorrect predictions. Then, you use `torch.mean` to calculate the average (mean) of these values. The result is added to the `accuracy` variable.  
   
  
**For example:**
```python
equals = [False, True, True]
accuracy += torch.mean(equals.type(torch.FloatTensor))
```
`equals.type(torch.FloatTensor)` converts boolean to float, resulting in `[0.0, 1.0, 1.0]`. `torch.mean` calculates the average (2/3). `accuracy` is incremented by 2/3.

**Key Differences:**
- The first approach calculates accuracy by summing the number of correct predictions directly.
- The second approach calculates accuracy by taking the mean of 1.0 (for correct predictions) and 0.0 (for incorrect predictions), effectively giving the ratio of correct predictions.

**Notes:**
- The first approach is commonly used when calculating accuracy in classification tasks.
- The second approach can be used if you want to obtain the average accuracy directly (a value between 0 and 1), which might be useful in some scenarios.
