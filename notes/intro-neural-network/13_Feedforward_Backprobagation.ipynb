# <span style="background-color: #FFFF99; color: #000000; font-weight: bold;">**Feedforward:**</span>  
[Watch Feedforward video](https://www.youtube.com/watch?v=hVCuvMGOfyY)

In a nutshell, feedforward is the process through which input data is passed through the neural network, layer by layer, to produce an output. Each layer performs a weighted sum of its inputs, applies an activation function, and passes the result to the next layer. This progression continues until the final layer produces the network's prediction. Feedforward is the forward flow of information from input to output without considering any feedback or adjustments.

# <span style="background-color: #FFFF99; color: #000000; font-weight: bold;">**Backpropagation:**</span>  
[Watch Backpropagation video](https://www.youtube.com/watch?v=1SmY3TZTyUk&t=1s)

Backpropagation is the learning algorithm used in training neural networks. In a nutshell, it involves two main steps:

1. **Forward Pass (Feedforward):**
   - Input data is passed through the network to make predictions.
   - The predicted output is compared to the actual output to calculate the error.

2. **Backward Pass (Backpropagation):**
   - The error is propagated backward through the network to calculate the gradients of the error with respect to the weights and biases.
   - These gradients are used to update the weights and biases in the direction that minimizes the error.
   - This process is repeated iteratively over multiple epochs to improve the model's performance.

In summary, feedforward is the process of making predictions, while backpropagation is the process of updating weights and biases based on the prediction error, aiming to improve the model's accuracy over time. Together, they form the basis of training neural networks for various tasks.

# Chain Rule  
[Watch Chain Rule video](https://www.youtube.com/watch?v=YAhIBOnbt54)  

- During the backward pass (backpropagation), the chain rule is used to calculate the gradients of the overall error with respect to the weights and biases.
- Starting from the output layer and moving backward, the chain rule is applied layer by layer.


In calculus, the chain rule is a fundamental concept that describes how to find the derivative of a composite function. For a composite function $F(g(x))$, the chain rule states that the derivative is the product of the derivative of the outer function $F'$ evaluated at the inner function $g(x)$ and the derivative of the inner function $g'(x)$.
